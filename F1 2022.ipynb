```python
# Libraries and Dataset

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
dataset = pd.read_csv('F1 2022.csv')
dataset.head(22)
```

```python
# X = Drivers ID, y = Position, Z = Constructors ID, H = Points

X = dataset.iloc[:,2:-4].values
y = dataset.iloc[:,:-6].values
Z = dataset.iloc[:,5:-1].values
H = dataset.iloc[:,6:].values
```

```python
# Driver Position Scatter

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth='0.5')
plt.scatter(X, y, color = 'blue')
plt.title('2022 Driver Position')
plt.xlabel('Driver ID')
plt.ylabel('Position')
```

```python
# Driver Position Barplot

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x='driver id', y='position', data=dataset)
plt.title('2022 Driver Position')
plt.xlabel('Driver ID')
plt.ylabel('Position')
```

```python
# Constructor Position Scatter

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth='0.5')
plt.scatter(Z, y, color = 'blue')
plt.title('2022 Constructor Position')
plt.xlabel('Constructor ID')
plt.ylabel('Position')
```

```python
# Constructor Position Barplot

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x='constructor', y='position', data=dataset)
plt.title('2022 Constructor Position')
plt.xlabel('Constructor')
plt.ylabel('Position')
```

```python
# Driver Points Scatter 

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth='0.5')
plt.scatter(H, X, color = 'blue')
plt.title('2022 Driver Points')
plt.xlabel('Points')
plt.ylabel('Driver ID')
```

```python
# Driver Points Barplot

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x='points', y='driver', data=dataset)
plt.title('2022 Driver Points')
plt.xlabel('Points')
plt.ylabel('Drivers')
```

```python
# Constructor Points Scatter 

plt.figure(figsize=(15,10))
plt.grid(color='green', linestyle='--', linewidth='0.5')
plt.scatter(H, Z, color = 'blue')
plt.title('2022 Constructor Points')
plt.xlabel('Points')
plt.ylabel('Constructor ID')
```

```python
# Constructor Points Barplot

f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x='points', y='constructor', data=dataset)
plt.title('2022 Constructor Points')
plt.xlabel('Points')
plt.ylabel('Constructors')
```
